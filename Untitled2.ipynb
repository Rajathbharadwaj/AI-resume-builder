```python
def parallel_processing(description):
    
    researcher = Agent(
    role="Tech Job Researcher",
    goal="Make sure to do amazing analysis on "
         "job posting with contrasting with the job applicant's resume",
    tools = [csv_search_tool, semantic_search_resume],
    verbose=True,
    backstory=(
        "As a Job Researcher, your prowess in "
        "navigating and extracting critical "
        "information from job postings."
        "Your skills help pinpoint the necessary "
        "qualifications and skills sought "
        "by employers, forming the foundation for "
        "effective application tailoring."
        "You master in profile matching and giving the"
        "profile matching percentage"
        )
    )

    res_matcher = Agent(
    role="Resume Matching Expert",
    goal="Ensure exceptional matching of resumes to job descriptions "
         "to aid recruiters and job applicants.",
    tools=[csv_search_tool, semantic_search_resume],
    verbose=True,
    backstory=(
        "As a Resume Matching Expert, your expertise lies in analyzing "
        "resumes and job descriptions with remarkable precision. "
        "You excel at identifying the most relevant skills, experiences, "
        "and qualifications that align with job requirements, "
        "thereby enhancing the chances of successful job placements. "
        "Your keen understanding of industry-specific terminologies and "
        "your ability to draw meaningful connections between candidate profiles "
        "and employer needs make you an invaluable asset in the recruitment process."
        )
    )

    # task
    research_task = Task(
    description=(
        "Analyze the job description provided "
        "to extract key skills, experiences, and qualifications "
        "required. Use the tools to gather content and identify "
        "and categorize the requirements."
        "Here's the job description ({job_description})"
    ),
    expected_output=(
        "A structured list of job requirements, including necessary "
        "skills, qualifications, and experiences."
    ),
    agent=researcher,
    )

    resume_matching_task = Task(
    description=(
        "Using the job description and resume provided, conduct a detailed analysis to "
        "match the resume with the job requirements. Utilize tools to enhance and adjust "
        "the resume content, ensuring it highlights the most relevant skills and experiences "
        "without fabricating any information. Update all sections, including the summary, "
        "work experience, skills, and education, to better reflect the candidate's fit for "
        "the job."
    ),
    expected_output=(
        "A job matching percentage along with the company the job description matches the most with "
        "The top 12 companies in highest to lowest percentage matching order"
    ),
    output_file="matches.json",
    context=[research_task],
    agent=res_matcher
    )


    job_application_crew = Crew(
    agents=[researcher,
            res_matcher],

    tasks=[research_task,
           resume_matching_task,],

    verbose=True
    )
    crew_inputs = {'job_description':description}

    result = job_application_crew.kickoff(inputs=crew_inputs)

    return result
    
```

```python
from joblib import Parallel, delayed
```

```python
processed_descriptions = Parallel(n_jobs=-1, prefer="threads")(delayed(parallel_processing)(desc) for desc in jd[0:2])
```
